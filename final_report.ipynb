### 4.2 The Baseline Model
  > We designed and implemented our own CNN model to use as a baseline in comparison to other popular deep learning networks for image classification. We used a combination of convolutional and max pooling layers with primarily relu activation as well as a few dropout layers to counter potential overfitting. Finally the model we designed concludes with several fully connected linear layers to condense how the model understands each image down to a single binary classification. For the CNN baseline, the final F1 score was 96 after using the weighted BCE Loss for 30 epochs.


### 4.3 VGG Model
  > Our first model tested is a VGG. Propsed in "Very Deep Convolutional Networks for Large-Scale Image Recognition" in 2014, this CNN based deep learning network architecture was motivated by "the deeper the networks, the better the results", and is widely used nowadays for transfer learning for other upstream machine learning stasks. We constructed and tested this model in "vgg.ipynb". In the file we first define several functions used for training and testing the model. We then constructed four models, two were initialized with standard pretrained weights and two were intialized without. We then loaded our pneumonia data, applied our transforms to the training set, and trained two models, one with standard pretraining and one without, on the pneumonia data. Next we loaded the COVID data and trained and tested all four models, one pretrained twice, one pretrained on only the pneumonia data, one with only the standard pretraining, and one with no pretraining at all, for a set number of epochs logging the results of each model after each epoch. Finally we showed the logged results and visualized these results with a plot. We found that while the untrained model clearly started off much worse achieving less than 85% F1 accuracy after the first epoch, it quickly caught up and achieved a 95% F1 accuracy after just ten epochs. ### 4.4 VIT Model
  > Our second model is a vision transformer, proposed in "An image is worth 16x16 words: Transformers for image recognition at scale", it follows the attention mechanism that was commonly used in natural language processing task. It split an image into patches and try to learn not only the representation of each patch, but also the importance of different patch. Finally it aggregate all of the weighted representation generated from patches, and send it to the classifier layer for classification. We explored this model in "vit_with_pretrain.ipynb". For vision transformer, the final F1 score was improved from 95 to 96 after using the weighted BCE Loss. And the covergence speed is the same (around 6 epochs).
